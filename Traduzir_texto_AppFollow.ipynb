```python
# importando pacotes pra manipula√ß√£o dos dados
import pandas
import csv
from googletrans import Translator
tradutor = Translator()
```

```python
# conectando Drive ao Colab
from google.colab import drive
drive.mount('/content/drive')
```

```python
# acessando a pasta onde est√£o os dados

lista = ['store','date','rating','title','internal_id','review_id','country','content','author','lang']

listacols = ['store','date','rating','title','internal_id','review_id','country','content','author','lang']

df = pandas.read_csv("/content/drive/MyDrive/Colab Notebooks/appfollow.csv", sep=",", usecols= lista)
df.columns = listacols
df.head(15)
```

```python
# opcional para evitar m√° formata√ß√£o do arquivo
# removendo line breaks do texto na coluna content
df['content']  = df['content'].replace('\\n',' ', regex=True)
df['content']  = df['content'].replace('\\n\\n', ' ', regex=True)
df.head(20)
```

```python
# agrupando colunas title e content para an√°lise completa da review
df['review'] = df['title'].fillna('') + " " + df['content'].fillna('')
df.head(20)
```

```python
# opcional para evitar m√° formata√ß√£o do arquivo
# removendo line breaks do texto agrupado
df['review']  = df['review'].replace('\\n',' ', regex=True)
df['review']  = df['review'].replace('\\n\\n', ' ', regex=True)
df.head(20)
```

```python
'''
essa fun√ß√£o detecta se os dados da coluna s√£o "en" de ingles 
e traduz para portugues
'''

def traduzirEN(value):
  if str(tradutor.detect(value))[14:16] == "en":
    td = tradutor.translate(value, src="en", dest="pt")
    return td.text
```

```python

'''
essa fun√ß√£o detecta se os dados da coluna s√£o "es" de espanhol 
e traduz para portugues
'''
def traduzirES(value):
  if str(tradutor.detect(value))[14:16] == "es":
    td = tradutor.translate(value, src="es", dest="pt")
    return td.text
```

```python
# aplicando tradu√ß√£o EN
df['traduzidoEN'] = df.review.map(traduzirEN)
df[['store','date','rating','title','internal_id','review_id','country','content','author','lang','review','traduzidoEN']]
```

```python
# aplicando tradu√ß√£o ES
df['traduzidoES'] = df.review.map(traduzirES)
df[['store','date','rating','title','internal_id','review_id','country','content','author','lang','review','traduzidoEN','traduzidoES']]
```

```python
# inserindo valores onde est√° nulo
df['traduzidoEN'] = df['traduzidoEN'].fillna(df['traduzidoES'])
df.head(20)
```

```python
# inserindo valores onde est√° nulo
df['traduzidoEN'] = df['traduzidoEN'].fillna(df['review'])
df.head(20)
```

```python
# tornando uma coluna s√≥ com textos em PT-BR
df['traduzido'] = df['traduzidoEN']
df.head(20)
```

```python
# filtrando colunas que ser√£o utilizadas
df = df[['store','date','rating','title','internal_id','review_id','country','content','author','lang','review','traduzido']]
```
